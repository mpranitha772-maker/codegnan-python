```python
del len
n = int(input())
temp = n
result = 0
length = len(str(n))

while n > 0:
    rem = n % 10
    result += rem ** length
    n //= 10

if result == temp:
    print(f"{temp} is an Armstrong number")
else:
    print(f"{temp} is not an Armstrong number")
```
